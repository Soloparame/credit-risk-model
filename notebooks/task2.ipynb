```python
# Replace this path with the actual one
df = pd.read_csv('../data/raw/data.csv')
df.shape
```

```python
numerical = df.select_dtypes(include=np.number).columns.tolist()

for col in numerical:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
if 'df' in globals():
    categorical = df.select_dtypes(include='object').columns.tolist()

    df_sample = df.sample(10000, random_state=42) if len(df) > 10000 else df

    for col in categorical:
        if df_sample[col].nunique() > 30:
            continue
        plt.figure(figsize=(6, 4))
        sns.countplot(data=df_sample, x=col, order=df_sample[col].value_counts().index)
        plt.title(f'Category distribution: {col}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
else:
    print("DataFrame 'df' is not defined. Please run the cell where 'df' is created.")
```

```python
# Ensure 'df' is defined before running this cell
if 'df' in globals():
	plt.figure(figsize=(10, 8))
	corr = df[numerical].corr()
	sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
	plt.title("Correlation Matrix")
	plt.show()
else:
	print("DataFrame 'df' is not defined. Please run the cell where 'df' is created.")
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

plt.figure(figsize=(8, 4))
sns.barplot(x=missing.values, y=missing.index)
plt.title("Missing Values by Feature")
plt.xlabel("Count")
plt.show()
```

```python
for col in numerical:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```
